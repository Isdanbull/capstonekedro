```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
def score_f(row):
    return (3*row.kills) + (2*row.assists) + (0.02*row['total cs']) + (2*row.firstbloodkill) + (2*(row.triplekills-row.quadrakills)) + (4*(row.quadrakills-row.pentakills)) + (7*row.pentakills) - row.deaths
```

```python
raw = pd.read_csv('ml_data.csv', index_col = 'Unnamed: 0')
```

```python
raw.head()
```

```python
western = ['LCS', 'LEC', 'NA LCS', 'EU LCS', 'LCK']
data = raw[raw.league.isin(western)].copy()
data.league.value_counts()
```

```python
data.loc[:,'date'] = pd.to_datetime(data.date)
data['score'] = data.apply(lambda row: score_f(row), axis=1)
```

```python
data.dtypes
```

```python
data['week'] = data.date.dt.isocalendar()['week']
```

```python
data.groupby(['league', 'year', 'week', 'position'])[['score']].max()
```
