```python
import pandas as pd
import numpy as np

import datetime as dt
import itertools

import processing_func

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_22 = pd.read_csv('2022_LoL_esports_match_data_from_OraclesElixir_20220324.csv')
df_21 = pd.read_csv('2021_LoL_esports_match_data_from_OraclesElixir_20220324.csv')
df_20 = pd.read_csv('2020_LoL_esports_match_data_from_OraclesElixir_20220324.csv')
df_19 = pd.read_csv('2019_LoL_esports_match_data_from_OraclesElixir_20220324.csv')
df_18 = pd.read_csv('2018_LoL_esports_match_data_from_OraclesElixir_20220324.csv')
df_17 = pd.read_csv('2017_LoL_esports_match_data_from_OraclesElixir_20220324.csv')
df_16 = pd.read_csv('2016_LoL_esports_match_data_from_OraclesElixir_20220324.csv')
df_15 = pd.read_csv('2015_LoL_esports_match_data_from_OraclesElixir_20220324.csv')
df_14 = pd.read_csv('2014_LoL_esports_match_data_from_OraclesElixir_20220324.csv')
frames = [df_22, df_21, df_20, df_19, df_18, df_17, df_16, df_15, df_14]
```

```python
#raw = df_20
raw = pd.concat(frames, axis=0, join='outer', ignore_index=True)
raw.shape
```

```python
raw.head()
```

```python
data = raw
data.dropna(subset=['playerid'], inplace=True)
leagues = ['LPL','LCK', 'LCS', 'LEC']
data = data[data.league.isin(leagues)].copy()
```

```python
def score_f(row):
    return (3*row.kills) + (2*row.assists) + (0.02*row['total cs']) + (2*row.firstbloodkill) + (2*(row.triplekills-row.quadrakills)) + (4*(row.quadrakills-row.pentakills)) + (7*row.pentakills) - row.deaths

data['score'] = data.apply(lambda row: score_f(row), axis=1)
```

```python
data['side_pos'] = data['side'] + data['position']
data['key'] = data['league'] + data['date'] +data['gameid']
data.groupby(['key', 'side_pos']).head()
```

```python
days_df = data.pivot(index='key', columns='side_pos', values='playerid')
days_df.head()
```

```python
sides = list(data.side.unique())
positions = list(data.position.unique())
s_p = list(itertools.product(sides, positions))

days = data.gameid.unique()
days_df = pd.DataFrame(days, columns=['gameid'])
days_df.sort_values('gameid')

print(days_df.gameid.value_counts().sort_values(ascending=False).head())

play_cols = []

for i in s_p:
    temp = data[(data.side == i[0]) & (data.position == i[1])][['gameid', 'playerid']]
    days_df = pd.merge(days_df, temp, on='gameid', how='left')
    days_df.rename(columns={'playerid':i[0]+i[1]}, inplace=True)
    print(days_df.gameid.value_counts().sort_values(ascending=False).head())
    play_cols.append(i[0]+i[1])
days_df.head()
```

```python
days_df.date.value_counts().sort_values(ascending=False).head(10)
```

```python
aggs= data.groupby(['playerid', 'date'])[['kills', 'deaths', 'assists', 'total cs', 'score',
                                           'firstbloodkill', 'triplekills', 'quadrakills', 'pentakills']].sum()

cols = list(aggs.columns)
renamer = {}
for col in cols:
    renamer[col] = 'average_'+col
renamer['total cs'] = 'average_cs'

games = data.groupby(['playerid', 'date'])[['playerid']].count()
aggs = pd.merge(aggs, games, left_index=True, right_index=True, how='inner')

renamer['playerid'] = 'total_games'

aggs = aggs.groupby(level=0).cumsum().rename(columns=renamer)
aggs = aggs.groupby(level=0).shift(1)

for i in renamer.values():
    if i != 'total_games':
        aggs[i] = aggs[i]/aggs['total_games']
        
aggs['total_games'].fillna(0, inplace=True)
aggs.fillna(-1, inplace=True)
agg_cols = list(aggs.columns)
aggs.reset_index(inplace=True)
```

```python
days_df = days_df.reindex(columns = list(days_df.columns)+agg_cols+['score'])
```

```python
days_df.columns
```

```python
scores = data.groupby(['playerid', 'date'])[['score']].sum().reset_index()
```

```python
days_df.groupby('date')[['Bluetop']].count().sort_values(by='Bluetop', ascending=False).head()
```

```python

for col in play_cols:
    
    days_df = pd.merge(days_df, aggs, left_on=[col, 'date'], right_on=['playerid', 'date'],
                       how='left', suffixes=('','_'+col )).drop(columns='playerid')
    
    days_df = pd.merge(days_df, scores,
                       left_on=[col, 'date'], right_on=['playerid', 'date'],
                       how='left', suffixes=('','_'+col )).drop(columns='playerid')
    
drop_cols = agg_cols + play_cols +['score']    
days_df.drop(columns=drop_cols, inplace=True)

days_df.groupby('date')[['average_kills_Bluetop']].count().sort_values(by='average_kills_Bluetop', ascending=False).head()
```
